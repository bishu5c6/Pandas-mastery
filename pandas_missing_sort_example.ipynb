```python
import pandas as pd
import numpy as np

popularity = {"Exam 1":[50,47,44,53,42],
             "Exam 2":[60,57,46,59,47],
             "Exam 3":[80,77,65,72,63],
             "Exam 4":[81,83,75,79,87]}
dataframe1 = pd.DataFrame(popularity, index = ['School A','School B','School C','School D','School E'])

dataframe1
```

```python
dataframe2 = dataframe1[dataframe1<70]
dataframe2
```

```python
dataframe2.dropna()
```

```python
dataframe2.dropna(axis=1)
```

```python
dataframe2.dropna(axis=1, how='all')
# drop the column which have all values as NaN
# dataframe2.dropna(axis=1, how='any')
# drop the column which have any one value as NaN
```
